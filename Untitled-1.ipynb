```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv(r"D:\다운로드\Crash_Reporting_-_Drivers_Data.csv")

# 10,000건 랜덤 샘플링
df = df.sample(n=10000, random_state=42)

# 날짜 컬럼 변환
df["Crash Date/Time"] = pd.to_datetime(df["Crash Date/Time"], errors="coerce")

# 요일 및 시간 추출
df["Hour"] = df["Crash Date/Time"].dt.hour
df["Weekday"] = df["Crash Date/Time"].dt.day_name()
```

```python
import folium
from folium.plugins import HeatMap

# 위도와 경도 결측치 제거
map_data = df[["Latitude", "Longitude"]].dropna()

# 너무 많으면 지도 로딩 느리니 5000건으로 제한
map_data = map_data.sample(n=5000, random_state=42)

# 기본 지도 객체 생성 (중심 좌표는 평균값)
base_map = folium.Map(
    location=[map_data["Latitude"].mean(), map_data["Longitude"].mean()],
    zoom_start=10
)

# HeatMap 추가
heat_data = list(zip(map_data["Latitude"], map_data["Longitude"]))
HeatMap(heat_data, radius=8, blur=15).add_to(base_map)

# 결과 지도 출력
base_map
```

```python
# 전처리 (책임, 요일 숫자화, 범주형 더미화 등)
df["Driver At Fault"] = df["Driver At Fault"].map({"YES": 1, "NO": 0})
df["Weekday_num"] = df["Crash Date/Time"].dt.weekday
df["Weather"] = df["Weather"].str.upper().str.strip()
df = df[df["Weather"].isin(df["Weather"].value_counts().nlargest(5).index)]

# 범주형 변수 인코딩
df_model = pd.get_dummies(df[["Hour", "Weekday_num", "Speed Limit", "Vehicle Year", "Weather", "Driver At Fault"]].dropna(), drop_first=True)

# 분류
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X = df_model.drop("Driver At Fault", axis=1)
y = df_model["Driver At Fault"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

coords = df[["Latitude", "Longitude"]].dropna()
coords_scaled = StandardScaler().fit_transform(coords)

# 5개 지역 군집
kmeans = KMeans(n_clusters=5, random_state=42)
df["RegionCluster"] = kmeans.fit_predict(coords_scaled)

import folium
from folium.plugins import MarkerCluster

cluster_map = folium.Map(location=[coords["Latitude"].mean(), coords["Longitude"].mean()], zoom_start=10)
marker_cluster = MarkerCluster().add_to(cluster_map)

for _, row in df.sample(n=1000).iterrows():
    folium.CircleMarker(
        location=(row["Latitude"], row["Longitude"]),
        radius=3,
        color=f"#{hex(100 + row['RegionCluster'] * 30)[2:]}FFAA",
        fill=True,
        fill_opacity=0.7
    ).add_to(marker_cluster)

cluster_map
```
